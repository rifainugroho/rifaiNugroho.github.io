### Menampilkan Koordinat Node dan Nama Node

```python
import folium
from haversine import haversine

# definisikan koordinat node
A = -6.81458,107.62306
B = -6.81495,107.61428
C = -6.81752,107.62242
D = -6.82171,107.63139
E = -6.82271,107.62295
F = -6.82675,107.62552
G = -6.82872,107.63081
H = -6.84915,107.61691
I = -6.86230,107.59622
J = -6.86638,107.62102
K = -6.88335,107.60498
L = -6.88704,107.61508

# buat kamus untuk memudahkan pencarian koordinat node berdasarkan nama node
nodes = {
    'A': A,
    'B': B,
    'C': C,
    'D': D,
    'E': E,
    'F': F,
    'G': G,
    'H': H,
    'I': I,
    'J': J,
    'K': K,
    'L': L
}
```

### Menampilkan Jarak Node A-C Beserta Mapnya

```python
# minta input node awal dari pengguna (hanya boleh A)
while True:
    node_a = input("Inputkan nama node awal (A): ")
    if node_a == 'A':
        break
    print("Node awal yang diinputkan salah! Hanya boleh menginputkan node A dengan huruf kapital.")

# minta input node akhir dari pengguna (hanya boleh C)
while True:
    node_c = input("Inputkan nama node akhir (C): ")
    if node_c == 'C':
        break
    print("Node akhir yang diinputkan salah! Hanya boleh menginputkan node C dengan huruf kapital.")

# hitung jarak antara dua node berdasarkan koordinatnya
jarak = haversine(nodes[node_a], nodes[node_c])

# tampilkan jarak
print(f"Jarak Terpendek {node_a}-{node_c} adalah {jarak:.2f} km.")

# Define the map center
map_center = (-6.83333, 107.61667)

# Create a map object
m = folium.Map(location=map_center, zoom_start=12)

# Add markers for each node
for node_name, node_location in nodes.items():
    if node_name == 'A':
        folium.Marker(location=node_location, tooltip=node_name, icon=folium.Icon(color='green')).add_to(m)
    elif node_name == 'L':
        folium.Marker(location=node_location, tooltip=node_name, icon=folium.Icon(color='red')).add_to(m)
    else:
        folium.Marker(location=node_location, tooltip=node_name).add_to(m)
    
# Get the coordinates of the starting and ending nodes
start_location = nodes[node_a]
end_location = nodes[node_c]

# Create a polyline between the starting and ending nodes
folium.PolyLine(locations=[start_location, end_location], color='red').add_to(m)

# Prompt the user to choose whether to display the map or not
tampilkan_map = input("Tampilkan map? (y/n): ")

# Display the map if the user chooses to
if tampilkan_map == 'y':
    display(m)
elif tampilkan_map == 'n':
    print("Map tidak ditampilkan.")
else:
    print("Pilihan tidak valid. Map tidak ditampilkan.")
```
