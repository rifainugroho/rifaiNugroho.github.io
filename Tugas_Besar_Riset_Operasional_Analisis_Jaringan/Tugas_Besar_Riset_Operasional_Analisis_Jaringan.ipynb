```python
import folium

# definisikan koordinat node
A = -6.81458,107.62306
B = -6.81495,107.61428
C = -6.81752,107.62242
D = -6.82171,107.63139
E = -6.82271,107.62295
F = -6.82675,107.62552
G = -6.82872,107.63081
H = -6.84915,107.61691
I = -6.86230,107.59622
J = -6.86638,107.62102
K = -6.88335,107.60498
L = -6.88701,107.61552

# buat objek peta
peta = folium.Map(location=A, zoom_start=12)

# tambahkan marker untuk setiap node
nodes = [A, B, C, D, E, F, G, H, I, J, K, L]
node_names = ['Node A',
              'Node B',
              'Node C',
              'Node D',
              'Node E',
              'Node F',
              'Node G',
              'Node H',
              'Node I',
              'Node J',
              'Node K',
              'Node L']

for i in range(len(nodes)):
    if i == 0: # node A
        folium.Marker(location=nodes[i], tooltip=node_names[i], icon=folium.Icon(color='green')).add_to(peta)
    elif i == len(nodes)-1: # node L
        folium.Marker(location=nodes[i], tooltip=node_names[i], icon=folium.Icon(color='red')).add_to(peta)
    elif i > 0 and i < 11: # node B-K
        folium.Marker(location=nodes[i], tooltip=node_names[i], icon=folium.Icon(color='blue')).add_to(peta)
    else:
        folium.Marker(location=nodes[i], tooltip=node_names[i]).add_to(peta)

# hubungkan setiap node dengan jalur
# definisikan urutan node dalam jalur
jalur = [A, C, B, I, K, L, J, G, D, C, E, F, H, K, I, E, F, G, J, H]

# tambahkan jalur ke peta
folium.PolyLine(locations=jalur, color='black').add_to(peta)

# tampilkan peta
peta
```

### Menampilkan Node, Jarak Antar Node Beserta Nama Jalannya

```python
from geopy.distance import geodesic
import pandas as pd

# Koordinat
locations = {
    'A': (-6.81458, 107.62306),
    'B': (-6.81495, 107.61428),
    'C': (-6.81752, 107.62242),
    'D': (-6.82171, 107.63139),
    'E': (-6.82271, 107.62295),
    'F': (-6.82675, 107.62552),
    'G': (-6.82872, 107.63081),
    'H': (-6.84915, 107.61691),
    'I': (-6.86230, 107.59622),
    'J': (-6.86638, 107.62102),
    'K': (-6.88335, 107.60498),
    'L': (-6.88701, 107.61552)
}

# Definisi jalan antar titik
routes = [
    ('A', 'C', 'Jl. Panorama'),
    ('C', 'B', 'Jl. Grand Hotel'),
    ('B', 'I', 'Jl. Raya Lembang'),
    ('I', 'K', 'Jl. Dr. Setiabudi'),
    ('K', 'L', 'Jl. Siliwangi'),
    ('C', 'E', 'Jl. Punclut'),
    ('E', 'I', 'Jl. Cijengkol'),
    ('E', 'F', 'Jl. Punclut'),
    ('F', 'G', 'Jl. Pager Wangi'),
    ('F', 'H', 'Jl. Punclut'),
    ('H', 'K', 'Jl. Punclut'),
    ('H', 'J', 'Jl. Citra Green'),
    ('C', 'D', 'Jl. Kayu Ambon'),
    ('D', 'G', 'Jl. Buka Nagara'),
    ('G', 'J', 'Jl. Dago Giri'),
    ('J', 'L', 'Jl. Ir. H. Djuanda')
]

# Hitung jarak dan buat DataFrame
data = []
for route in routes:
    start, end, road_name = route
    distance = geodesic(locations[start], locations[end]).km
    data.append((f'{start}-{end}', distance, road_name))

df = pd.DataFrame(data, columns=['Node', 'Jarak (km)', 'Nama Jalan'])

# Memformat kolom 'Jarak (km)' menjadi 2 angka di belakang koma
df['Jarak (km)'] = df['Jarak (km)'].round(2)

# Tampilkan tabel
df
```

```python
import folium
from geopy.distance import geodesic

# Definisikan urutan jalur
jalur_list = [
    ['A', 'C', 'B', 'I', 'K', 'L'],
    ['A', 'C', 'E', 'I', 'K', 'L'],
    ['A', 'C', 'E', 'F', 'H', 'K', 'L'],
    ['A', 'C', 'E', 'F', 'H', 'J', 'L'],
    ['A', 'C', 'E', 'F', 'G', 'J', 'L'],
    ['A', 'C', 'D', 'G', 'J', 'L']
]

# Definisikan koordinat node
nodes = {
    'A': (-6.81458, 107.62306),
    'B': (-6.81495, 107.61428),
    'C': (-6.81752, 107.62242),
    'D': (-6.82171, 107.63139),
    'E': (-6.82271, 107.62295),
    'F': (-6.82675, 107.62552),
    'G': (-6.82872, 107.63081),
    'H': (-6.84915, 107.61691),
    'I': (-6.86230, 107.59622),
    'J': (-6.86638, 107.62102),
    'K': (-6.88335, 107.60498),
    'L': (-6.88701, 107.61552)
}

# Definisikan kamus nama jalan berdasarkan node awal dan akhir
nama_jalan_map = {
    ('A', 'C'): 'Jl. Panorama',
    ('C', 'B'): 'Jl. Grand Hotel',
    ('B', 'I'): 'Jl. Raya Lembang',
    ('I', 'K'): 'Jl. Dr. Setiabudi',
    ('K', 'L'): 'Jl. Siliwangi',
    ('C', 'E'): 'Jl. Punclut',
    ('E', 'I'): 'Jl. Cijengkol',
    ('E', 'F'): 'Jl. Punclut',
    ('F', 'G'): 'Jl. Pager Wangi',
    ('F', 'H'): 'Jl. Punclut',
    ('H', 'K'): 'Jl. Punclut',
    ('H', 'J'): 'Jl. Citra Green',
    ('C', 'D'): 'Jl. Kayu Ambon',
    ('D', 'G'): 'Jl. Buka Nagara',
    ('G', 'J'): 'Jl. Dago Giri',
    ('J', 'L'): 'Jl. Ir. H. Djuanda'
}

import folium
from geopy.distance import geodesic

# ... (previous code for defining nodes, jalur_list, and nama_jalan_map)

# Fungsi untuk menghitung total jarak suatu jalur
def calculate_total_distance(jalur):
    total_distance = 0
    for i in range(len(jalur) - 1):
        start = jalur[i]
        end = jalur[i + 1]
        if start in nodes and end in nodes:
            distance = geodesic(nodes[start], nodes[end]).kilometers
            total_distance += distance
    return total_distance

# Cari jalur terpendek dan nilai terkecil
shortest_distance = float('inf')
shortest_path = []

for jalur in jalur_list:
    current_distance = calculate_total_distance(jalur)
    if current_distance < shortest_distance:
        shortest_distance = current_distance
        shortest_path = jalur

# Buat objek peta
peta = folium.Map(location=nodes['A'], zoom_start=12)

# Tambahkan marker untuk setiap node
for node, coordinates in nodes.items():
    icon_color = 'green' if node == 'A' else ('red' if node == 'L' else 'blue')
    folium.Marker(location=coordinates, tooltip=f'Node {node}', icon=folium.Icon(color=icon_color)).add_to(peta)

# Hubungkan setiap node dengan jalur dan tampilkan nama jalan
for i in range(len(shortest_path) - 1):
    start_node = shortest_path[i]
    end_node = shortest_path[i + 1]
    start_coordinates = nodes[start_node]
    end_coordinates = nodes[end_node]
    jalan = nama_jalan_map.get((start_node, end_node), f'Jl. {start_node}-{end_node}')
    folium.PolyLine(locations=[start_coordinates, end_coordinates], color='black', tooltip=jalan).add_to(peta)

print("Jalur terdekat:")
print(" -> ".join(shortest_path))
print("Total jarak: {:.2f} kilometer".format(shortest_distance))
print("Nama jalan yang dilalui:")
for i in range(len(shortest_path) - 1):
    start_node = shortest_path[i]
    end_node = shortest_path[i + 1]
    jalan = nama_jalan_map.get((start_node, end_node), f'Jl. {start_node}-{end_node}')
    print(f"{i + 1}. {jalan}")
# Tampilkan peta
peta
```
